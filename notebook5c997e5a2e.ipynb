# Sleep Health and Lifestyle – Data Visualization Project


## 1. Introduction
This project explores how sleep duration and quality relate to stress levels,
physical activity, and lifestyle factors using data visualization.



## 2. Dataset Overview
Source: Kaggle – Sleep Health and Lifestyle Dataset



## 3. Exploratory Data Analysis


**Plot 1: Sleep Duration Distribution **

```python
import matplotlib.pyplot as plt

plt.hist(df['Sleep Duration'], bins=10)
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Number of People')
plt.title('Sleep Duration Distribution')
plt.show()
```

Insight: More number of people sleep for 8 hours a day

Plot 2: Stress Level VS Sleep duration

```python
import matplotlib.pyplot as plt

plt.scatter(df['Stress Level'], df['Sleep Duration'])
plt.xlabel('Stress Level')
plt.ylabel('Sleep Duration (hours)')
plt.title('Stress Level vs Sleep Duration')
plt.show()
```

**Insight:** Higher stress levels are associated with shorter sleep duration.


Plot 3: Average sleep duration by gender 

```python
avg_sleep_gender = df.groupby('Gender')['Sleep Duration'].mean()

avg_sleep_gender.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Average Sleep Duration (hours)')
plt.title('Average Sleep Duration by Gender')
plt.show()
```

**Insight:** Sleep duration varies slightly by gender, though overall patterns remain similar.


Plot 4: Physical activity vs sleep

```python
plt.scatter(df['Physical Activity Level'], df['Quality of Sleep'])
plt.xlabel('Physical Activity Level')
plt.ylabel('Quality of Sleep')
plt.title('Physical Activity vs Quality of Sleep')
plt.show()
```

**Insight:** Higher physical activity levels tend to be associated with better sleep quality.


Plot 5 : Sleep disorder Distribution

```python
df['Sleep Disorder'].value_counts().plot(kind='bar')
plt.xlabel('Sleep Disorder Type')
plt.ylabel('Number of People')
plt.title('Distribution of Sleep Disorders')
plt.show()
```

**Insight:** Most individuals do not report sleep disorders, with insomnia and sleep apnea being the most common among reported cases.


Plot 6 : Sleep Duration by Occupation 

```python
df.groupby('Occupation')['Sleep Duration'].mean().sort_values().plot(kind='barh')
plt.xlabel('Average Sleep Duration (hours)')
plt.title('Average Sleep Duration by Occupation')
plt.show()
```

**Insight:** Sleep duration varies across occupations, possibly due to differences in work stress and schedules.
